```python
import os
import pandas as pd
import zipfile
from docxtpl import DocxTemplate


#Temporary Names
quarterly_file_name_template   = 'Q1 2016-17 Pacific Grove TEMPLATE.docx'
quarterly_file_name_target   = 'Q1 2016-17 Pacific Grove FINAL.docx'
data_source_file_name = 'scorecard_key_value_jurisdictions.csv'

#################################################################################
# Template & Data locations

template_location = os.path.join('..','..','reports','templates')
target_location   = os.path.join('..','..','reports','finals')
data_location     = os.path.join('..','..','data','processed')


#################################################################################
#Read in data and templates

data_source_file_path = os.path.join(data_location, data_source_file_name)
key_value_store = pd.read_csv(data_source_file_path)
keys = key_value_store['key'].tolist()
values = key_value_store['value'].tolist()
key_values = dict(zip(keys,values))


template_file_path       = os.path.join(template_location, quarterly_file_name_template)
target_final_report_file = os.path.join(target_location, quarterly_file_name_target)

doc = DocxTemplate(template_file_path)
doc.render(key_values)
doc.save("test_doc.docx")
```

```python

templateDocx = zipfile.ZipFile(template_file_path)
newDocx = zipfile.ZipFile(target_final_report_file, "a")

# Eventually we will access all files in the template subfolder and iterate through them

replaceText = {"XXXCLIENTNAMEXXX" : "Joe Bob", "XXXMEETDATEXXX" : "May 31, 2013"}

with open(templateDocx.extract("word/document.xml", "")) as tempXmlFile:
    tempXmlStr = tempXmlFile.read()

for key in replaceText.keys():
    tempXmlStr = tempXmlStr.replace(str(key), str(replaceText.get(key)))

with open("temp.xml", "w+") as tempXmlFile:
    tempXmlFile.write(tempXmlStr)

for file in templateDocx.filelist:
    if not file.filename == "word/document.xml":
        newDocx.writestr(file.filename, templateDocx.read(file))

newDocx.write("temp.xml", "word/document.xml")

templateDocx.close()
newDocx.close()
```
